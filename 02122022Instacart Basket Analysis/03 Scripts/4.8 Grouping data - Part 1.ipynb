### 03. Grouping data

#### 03.1 Create an object 

```python
df.groupby('product_name')
```

#### 03.2 Groupby with aggregate functions (e.g. mean)

```python
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
df.groupby('department_id')['order_number'].mean()
```

#### 03.3 Groupby with multiple aggregations

```python
#Performing multiple aggregations
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

### 04. Transform function

```python
# a new column max_order is created, to store the max. order nubmer by user_id
#transform () function is applied on the order_number column with np.max argument
#np is the numby library and the max()function is included in the numby to find the max value within a column.
# this tells python to trnasform the oder_number column by applying the max() function from the numpy library
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

### 05. Loc function to create a 'loyalty_flag' using 'max_order'

```python
# setting the conditions
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```
